```python
# Question 5. Three biggest pick up Boroughs
merge_df_pu = pd.merge(df, locations, left_on='PULocationID', right_on='LocationID', how='left')
merge_df_pu.head()
```

```python
merge_df_sept_19 = merge_df_pu.loc[(merge_df_pu['lpep_pickup_datetime'] >= '2019-09-18') & (merge_df_pu['lpep_pickup_datetime'] <= '2019-09-18 23:59:59')]
```

```python
merge_df_sept_19.drop(merge_df_sept_19[merge_df_sept_19['PULocationID'].isin([264, 265])].index, inplace=True)

merge_df_sept_19.drop(merge_df_sept_19[merge_df_sept_19['PULocationID'].isin([264, 265])].index, inplace=True)
```

```python
merge_df_sept_19.shape
```

```python
boroughs = {'borough1': ["Brooklyn", "Manhattan", "Queens"], 'borough2': ["Bronx", "Brooklyn", "Manhattan"], 'borough3': ["Bronx", "Manhattan", "Queens"], 'borough4': ["Brooklyn", "Queens", "Staten Island"]}

for borough_name, borough_list in boroughs.items():
    print(merge_df_sept_19.loc[merge_df_sept_19['Borough'].isin(borough_list), ["total_amount"]].sum())
```
